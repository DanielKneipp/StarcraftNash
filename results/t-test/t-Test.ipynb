# t-Test

## Reading the data

```python
import pandas as pd
import numpy as np

idx = list(range(1, 31))
# idx = list(range(1, 6))

data_stat = []
data_nonstat = []
for i in idx:
    df_stat = pd.read_csv('intermediate_stat/rep_' + str(i) + '_1515211501445.csv')
#     df_stat = pd.read_csv('dummy_data/rep_' + str(i) + '_1512096105723.csv')
    df_stat['rep'] = i
    data_stat.append(df_stat)
    
    df_nonstat = pd.read_csv('intermediate_nonstat/rep_' + str(i) + '_1515211218849.csv')
    df_nonstat['rep'] = i
    data_nonstat.append(df_nonstat)
```

### Filtering data

Removing the players: Fictitious(Stoch), NoRegret, Fictitious(Det)

```python
rem_players_lst = ['Fictitious(Stoch)', 'NoRegret', 'Fictitious(Det)']

# Getting the row ilocs
ilocs_to_rem = [data_stat[0].columns.get_loc(k) for k in rem_players_lst]

# Creating a map to remove the rows
iloc_filter = np.arange(len(data_stat[0]))
iloc_filter = ~np.isin(iloc_filter, ilocs_to_rem)

# Removing the rows and columns of each dataframe
for i in range(len(data_stat)):
    data_stat[i] = data_stat[i][iloc_filter]
    data_nonstat[i] = data_nonstat[i][iloc_filter]
    
    data_stat[i] = data_stat[i].drop(['Fictitious(Stoch)', 'NoRegret', 'Fictitious(Det)'], axis=1)
    data_nonstat[i] = data_nonstat[i].drop(['Fictitious(Stoch)', 'NoRegret', 'Fictitious(Det)'], axis=1)

data_stat[0].head()
```

### Combining dataframes

```python
data_stat_all = pd.concat(data_stat).reset_index(drop=True).applymap(lambda v: v if v != -1 else np.nan)
data_nonstat_all = pd.concat(data_nonstat).reset_index(drop=True).applymap(lambda v: v if v != -1 else np.nan)

# showing one of the dataframes
data_stat_all.head()
```

## Data statistics

### Average and std

```python
def calc_stats(df):
    df_avrg = df.mean()
    df_std = df.std()
    df_stats = pd.DataFrame([df_avrg, df_std], columns=df_avrg.index.values, index=['Average', 'Std'])
    return df_stats

data_stat_all_stats = calc_stats(100 - data_stat_all.drop('rep', 1))
data_nonstat_all_stats = calc_stats(100 - data_nonstat_all.drop('rep', 1))
```

## t-Test

### Calculating the means of each result (by repetition)

```python
data_stat_avrg = (100 - data_stat_all).groupby('rep').mean()
data_nonstat_avrg = (100 - data_nonstat_all).groupby('rep').mean()

data_stat_avrg.head()
```

### Calculatint t-Test for each agent

```python
from scipy.stats import ttest_ind

t_test = dict.fromkeys(data_stat_avrg.columns.values)

for agent in data_stat_avrg.columns:
    t_test[agent] = ttest_ind(data_stat_avrg[agent], data_nonstat_avrg[agent])
    
t_test_df = pd.DataFrame(t_test, index=['t-statistic', 'p-value'])
t_test_df.head()
```

## Saving the result

```python
t_test_df.to_csv('t-test.csv', encoding='utf-8')
```
